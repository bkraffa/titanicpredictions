Wrangle the data
---

```python
#dropping unnecessary features to speed up the training
train_df.drop(['Ticket','Cabin','PassengerId'], inplace=True, axis = 1)
test_df.drop(['Ticket','Cabin','PassengerId'], inplace=True, axis = 1)
combine = [train_df,test_df]
```

```python
#creating new feature from existing - 'name' - extracting the characters of the string until the first space on the feature name
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.', expand = False)

pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
#we can group the uncommon titles on a category named other
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')
    dataset['Title'] = dataset['Title'].replace('Mlle','Miss')
    dataset['Title'] = dataset['Title'].replace('Ms','Miss')
    dataset['Title'] = dataset['Title'].replace('Mme','Mrs')

train_df.groupby('Title').mean()
```

```python
#Then we can convert the categorical titles to ordinal
title_dict = {'Mr': 1,
            'Miss': 2,
            'Mrs': 3,
            'Master': 4,
            'Other': 5 }
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_dict) 
    dataset['Title'] = dataset['Title'].fillna(0)

train_df.head()
```

```python
#now we can also drop the name feature
train_df.drop(['Name'], axis = 1 , inplace = True)
train_df.head()
```

```python
#converting the categorical feature (sex) into ordinal
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'female':1,
                                        'male':0})
train_df.head()
```

```python
#now we should estimate or complete the feature with missing or null values, we'll start with the Age feature
#we will guess the missing values for age by using other correlated features like gender and pclass
graph = sns.FacetGrid(train_df, row = 'Pclass', col = 'Sex')
graph.map(plt.hist, 'Age', bins = 20)
graph.add_legend()
```

```python
#Lets prepare an empty array to contain the guessed age values for all the 6 pclass x gender combinations
guess_ages = np.zeros((2,3))
guess_ages
```

```python
#Now we iterate to get the median of each combination of pclass and sex, and use it to fill the missing value for each combination

for dataset in combine:
    for i in range (0,2):
        for j in range(0,3):
           guess_df = dataset[(dataset['Sex'] == i) & \
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()
           age_guess = guess_df.median()
           guess_ages[i,j] = int(age_guess/.5 +.5)*.5 #convert random age float to nearest .5 age

    for i in range (0,2):
        for j in range (0,3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()        
```

```python
#Checking to see how the ages split in 5 different bands (in absolut numbers not amount)
train_df['AgeBand'] = pd.cut(train_df['Age'],5)
train_df[['AgeBand','Survived']].groupby(['AgeBand'], as_index = False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
#Attributing a number to each of the agebands
for dataset in combine:
    dataset.loc[dataset['Age'] <= 16,'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32),'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48),'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64),'Age'] = 3
    dataset.loc[dataset['Age'] > 64,'Age'] = 4
train_df.head()
```

```python
train_df.drop(columns='AgeBand', inplace =True, axis = 1)
combine = [train_df,test_df]
train_df
```

```python
#Here we are aggregating the number of partners (parch) with simblings (SibSp) and the person itself (+1) to create a feature with the total number of members of the same family
for dataset in combine:
    dataset['FamilySize'] = dataset['Parch'] + dataset['SibSp'] + 1 
    dataset.drop(columns=['SibSp','Parch'],inplace=True,axis=1)    
train_df[['FamilySize','Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by=['Survived'],ascending = False)
```

```python
#Creating a feature 'isAlone' will help us to correlate the fact of being alone with the chance of surviving
for dataset in combine:
    dataset['isAlone'] = 0
    dataset.loc[dataset['FamilySize'] > 1, 'isAlone'] = 1
    dataset.drop('FamilySize',inplace=True,axis=1) #we can also drop family size as we've created isAlone
train_df[['isAlone','Survived']].groupby(['isAlone']).mean().sort_values(by=['Survived'],ascending=False)
#Its possible to see that the alone people had a higher survival mean rate
```

```python
#We can create a new feature multiplying age and the pclass, so in theory the lower number would be the people with the younger age and higher class 
for dataset in combine:
    dataset['AgeClass'] = dataset['Age'] * dataset['Pclass']

train_df[['Age','Pclass','AgeClass']]
```

```python
#Checking the embarked feature we can see that S is the most common port, so we'll apply it to the two missing values of the column
train_df.Embarked.describe()
```

```python
freq_port = 'S'
for dataset in combine:
    dataset['Embarked'].fillna(freq_port,inplace=True)
train_df[['Embarked','Survived']].groupby(['Embarked']).mean().sort_values(by=['Survived'],ascending=True)
#Its possible to see that the S port had the lower mean survival rate and C had the higher
```

```python
#Mapping the ports
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map({'S':0,
    'Q':1,
    'C':2}).astype(int)
train_df
```

```python
#Complete fare for the single missing value on the test DF using the mode
test_df['Fare'].fillna(test_df['Fare'].dropna().median(),inplace=True)
test_df
```

```python
#We can now create the bands for the fare, but as we did for the age we have to check the bands before
train_df['FareBand'] = pd.qcut(train_df['Fare'],4) #qcut divides into 4 quantiles for the bands
train_df[['FareBand','Survived']].groupby(['FareBand'],as_index=False).mean().sort_values(by=['FareBand'],ascending=True)
#We can see that the higher the band the higher the survival mean rate
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df.drop(['FareBand'], axis=1, inplace=True)
combine = [train_df, test_df]
    
train_df
```

```python
#And now both our datasets are ready
test_df.drop(columns=['Name'],inplace=True,axis=1)
test_df
```
